
# Part 1: Sorting algorithms

## Introduction

The objective of a sorting algorithm it to put a list of values in either ascending or descending order.
The efficiency of a sorting algorithm is evaluated by how little the process uses time and resources(memory). In todays computers memory is not as much of an issue as time. This project is meant to implement and then assess different sorting algorithms by comparing the duration they take to sort  lists of randomly generated values from length 10 up to 10 000

### Bubble Sort

#### Introduction


The way bubble sort function is by repeatably going through the list and comparing each element with the next. 
Swapping them when necessary until it passes through th full list without the need to swap any elements.
Meaning the list is fully sorted.

#### The Implementaion

````python
def bubble_sort(array):
    sorted=False
    array_size=len(array)
    while(not sorted):
        sorted=True
        for i in range(array_size-1):
            if(array[i]>array[i+1]):
                array[i+1],array[i]=array[i],array[i+1]
                sorted=False
    return array
````

#### Complexity



The algorithm uses two loops on inside the other. The inner loop runs $n$ times. The outer loop runs until the inner loop runs without effecting any changes to the array; 1 time in the best case and $n$ times in the worst case.
Since we are only interested in the worst case complexity, that gives and overall complexity of:
- $O(n^2)$

### Merge sort 


#### Introduction

Merge sort is a divide and conquer algorithm that was invented by the famous physicians John von Neumann in 1945. 
It functions by continually splitting the list into sub lists until they are one element.
Then merging each of the sub lists into a new sorted list.

#### The Implementation

````python
def merge_sort(array):
    length=len(array)
    if(length==1):
        return array
    ## Dividing the array 
    array1=merge_sort(array[:length//2])
    array2=merge_sort(array[length//2:])

    ## Merging 
    result_array=[]
    while(len(array1) and len(array2)):
        if(array1[0]<array2[0]):
           result_array.append(array1.pop(0))
        else:
           result_array.append(array2.pop(0))
    results_array=result_array+array1+array2
    return results_array
````

#### Complexity


The complexity function of this implementations is $T(n)=2T(n/2)+n$ using the master theorem we can conclude that:
- $O(n\log n)$

### Quick sort

#### Introduction

Quick sort is a sorting algorithm developer by Tony Hoare in 1959 and still in use today.
It is another divide and conquer algorithm it works by choosing a pivot The splitting the list into two depending on if the are greater or less then the pivot.

#### The implementation

````python
def quick_sort(array):
    if(len(array)==1):
        return array
    ## sorting the array with one less element
    popped=array.pop()
    sorted_array=quick_sort(array)
    
    ## if the popped element the biggest in the sorted array
    length=len(sorted_array)
    if(popped>=sorted_array[length-1]):
        sorted_array.append(popped)
        return sorted_array
    
    ## Using binary search to find the right placement
    high,low=length-1,0
    while(low<high):
        mid=(high+low)//2
        if(popped<=sorted_array[mid]):
            high=mid
        else:
            low=mid+1
    
    sorted_array.insert(low,popped)
    return sorted_array
````

#### Complexity

This algorithm uses recursivety passing an array of n-1 size. And Binary search to find the right placement for the popped element in the sorted array. Therefore the complexity is $T(n)=T(n-1)+\log(n)$ which gives the overall complexity of:
- $O(n\log n)$

Many implementations of this algorithm are $O(n^2)$ instead. The way we reduced the complexity was the use of binary search instead of linear search in the insertion.

### Insertion sort

#### Introduction

Insertion sort is a sorting algorithm that is so simple it could be implemented in three line in C++.
It function by going through the list and inserting each element into the part of the list is already ordered manner. 

#### Implementation

````python
def insertion_sort(array):
    length=len(array)
    for i in range(1,length):
        j=i
        while(j>0 and array[j-1]>array[j]):
            array[j-1],array[j]=array[j],array[j-1]
            j-=1
    return array
````

#### Complexity

The implementation uses outer loop that run n times. And an inner loop that run either 0 or n times. Which means this implementation has the complexity  of:

- $O(n^2)$

###  Heap sort

#### Introduction

Heap sort is an improved version of selection sort.
it is based on the binary heap data structure which is a binary tree where The descendent of each element are smaller then them.



#### Implementation

````python
def heapify(array,length , root):
    ## creates a heap assuming the two subtrees are already a heap
    largest = root
    left  = 2*root+1
    right = 2*root+2
    if left < length and array[largest] < array[left]:
        largest = left
    if right < length and array[largest] < array[right]:
        largest = right
    if largest != root:
        array[root], array[largest] = array[largest], array[root]
        heapify(array, length, largest)


def heap_sort(array):
    length = len(array)
    heapify(array,length,1)
    ## creating the heap from the bottom up (igonring the leafs)
    for i in range(length//2 - 1, -1, -1):
        heapify(array, length, i)

    # Turning the heap into a sorted array
    for i in range(length-1, 0, -1):
        array[i], array[0] = array[0], array[i]
        heapify(array, i, 0)
    return array
````

#### Complexity

Studying the complexity of this algorithm is a bit complicated. First heapify will either keep the root in it's place or push it down to the leafs which means a it will run a maximum of $\log_2(n)$. which mean it's $O(\log n)$.
For the sorting algorithm it runs the heapify function n/2 times in intializing the heap then n times in turning the heap back into a list. Therfore we can conclude The sorting algorithm of complexity:

- $O(n\log(n))$


### Selection sort

#### Introduction

Selection sort is another simple sorting algorithm.
It function by dividing the list into two parts. 
An orders section that is built from the left to right.
And an unordered section where the algorithm searches for the minimum value each time.


#### Implementation

````python

def selection_sort(array):
    length=len(array)
    if (length == 1):
        return array
    min_index = 0

    for i in range(length):
        if (array[i] < array[min_index]):
            min_index = i

    min = array.pop(min_index)
    array = selection_sort(array)
    array.insert(0, min)
    return array
````

#### Complexity

This algorithm usees one loop that runs n time. And a callback with and array of one less elemen as well as an insertion fucntion Which python implements in O(1).So the complexity is $T(n)=T(n-1)+n$  which gives us an overall complexity of:
- $O(n^2)$

## Comparing The Algorithms

To study the time complexity of these functions. we created a function that logs the durations these algorithms take to sort a randomly generated set of lists.
This function provides a few costumizable parameters. basically it will test start from and aray of size``` min_array_size``` then increase by ```step``` until ```max_array_size```.
For each array size we will use a number of samples defined by ```number_of_samples``` to get as accurate of a reading as possible. We also provide an estimation of the time left as well as a percentage of the arrays tested while the function is running.Finally the data is saved into the two arrays```array_sizes```and ```durations``` which is saved to a CSV fomat file. In this project we provide the data collected in the data folder.


````python
import pandas as pd
from random import random
from math import floor
import time

def run_algo(sort_name, arr):
    '''Runs the named algorithm with the provided array'''
    if sort_name == "bubble":
        bubble_sort(arr)
    if sort_name == "quick":
        quick_sort(arr)
    if sort_name == "merge":
        merge_sort(arr)
    if sort_name == "insertion":
        insertion_sort(arr)
    if sort_name == "heap":
        heap_sort(arr)
    if sort_name == "selection":
        selection_sort(arr)


MAX_ARRAY_SIZE = 7_000
MIN_ARRAY_SIZE = 0
MAX_NUMBER = 1_000
NUMBER_OF_SAMPLES = 5
STEP = 10


def study(max_array_size=MAX_ARRAY_SIZE,
          min_array_size=MIN_ARRAY_SIZE,
          max_number=MAX_NUMBER,
          number_or_samples=NUMBER_OF_SAMPLES,
          step=STEP,
          algo_name="bubble",
          array_sizes=[],
          durations=[],
          ):
    execution_time = 0
    for array_size in range(max_array_size, min_array_size, -step):
        for i in range(number_or_samples, 0, -1):
            start = time.time()
            ## Estimating the duration left
            print(f' {floor((max_array_size-array_size)*100/max_array_size)}%' +
                  f' ETA:{floor(number_or_samples*array_size*execution_time/step)}s', end='  \r')

            arr = [floor(max_number*random()) for _ in range(array_size)]
            run_algo(algo_name, arr)
            end = time.time()
            execution_time = end-start
            array_sizes.append(array_size)
            durations.append(execution_time)
````

### Example

We'll use Insertion Sort to demonstrate The study function. with array sizes up to 2000.

````python
times = {}
arr_sizes = []
durations = []
study(array_sizes=arr_sizes,
      durations=durations,
      min_array_size=0,
      step=10,
      max_array_size=2000,
      algo_name="insertion",
        )

times["array sizes"] = arr_sizes
times["durations"] = durations
df = pd.DataFrame(times)
print(df)
````

We will use the matplotlib package to visualise the data in a scatter diagram.

````python
import matplotlib.pyplot as plt

fig, axes = plt.subplots()
plt.rcParams["figure.figsize"] = (20,10)
axes.scatter(df["array sizes"], df["durations"],label=f'{algo_name}',sizes=[2])
plt.xlabel('Array size')
plt.ylabel('Executation time (ms)')
plt.legend(loc=2, prop={'size': 20},markerscale=6)
plt.show()
````

# Part 2: Matrix Multiplication

## Introduction

Matrix mulitplication is a very central operation in many numerical algorithms. The implementation of the mathematical defintion provides an algorithm that is of $O(n^2)$. However mathematicians and computer scientist have been able to develop algorithms that reduced that complexity. In 1969 Volger Strasen published his algorithm that reduced the copmlixity to $O(n^{2.373})$.

In this section we will impolement both algorthms and compare them. as well as the algorithm implemented in the python package numpy.

### Naive implementation

The naive matrix multiplication algorithm uses three loop each one runs n times. Giving the overall complixity of O(n^2).

````python
def classic_matrix_mult(A, B):
    n=len(A)
    R=[[0]*n]*n
    for i in range(n):
        for j in range(n):
            for k in range(n):
                R[i][j] += A[i][k]*B[i][k]
    return R
````

### Strassens algorithm

To implement Strassens algorithm, first we need a few helper functions.```add```,```substract``` and ```divide```.
The allow to do basic arithmetics on matrixes. The last one will devide the matrix into 4 sub matrixes.

````python
def add(A, B):
    n=len(A)
    R=[[0]*n]*n
    return[[A[i][j]+B[i][j] for j in range(n)] for i in range(n)]
        
def substract(A, B):
    n=len(A)
    return[[A[i][j]-B[i][j] for j in range(n)] for i in range(n)]


def divide(A):
    n = len(A)
    m = n // 2
    A11 = [[A[i][j] for j in range(m)] for i in range(m)]
    A12 = [[A[i][j] for j in range(m)] for i in range(m, n)]
    A21 = [[A[i][j] for j in range(m, n)] for i in range(m)]
    A22 = [[A[i][j] for j in range(m, n)] for i in range(m, n)]
    return A11,A12,A21,A22
````

Then we will implement the actual algorithm. 

````python
def strassen_matrix_mult(A, B,threshehold=512):
    if len(A) <=threshehold:
        C =classic_matrix_mult(A,B)

    else:
        print(len(A))
        A11, A12, A21, A22 = divide(A)
        B11, B12, B21, B22 = divide(B)

        s1 = strassen_matrix_mult(A11, substract(B12, B22))
        s2 = strassen_matrix_mult(add(A11, A12), B22)
        s3 = strassen_matrix_mult(add(A21, A22), B11)
        s4 = strassen_matrix_mult(A22, substract(B21, B11))
        s5 = strassen_matrix_mult(add(A11, A22), add(B11, B22))
        s6 = strassen_matrix_mult(substract(A12, A22), add(B21, B22))
        s7 = strassen_matrix_mult(substract(A11, A21), add(B11, B12))

        C11 = add(substract(add(s5, s4), s2), s6)
        C12 = add(s1, s2)
        C21 = add(s3, s4)
        C22 = substract(substract(add(s1, s5), s3), s7)
        print(len(A))

        C = []
        for i in range(len(C12)):
            C.append(C11[i] + C12[i])
        for i in range(len(C22)):
            C.append(C21[i] + C22[i])

    return C
````

We chose a relativly high number to switch the naive algorithm. This is because while strassens algorithm reduces the number of multiplications The overhead of array allocation, additions and substractions outweighs it's benefits in small enough matrixes.
